#Import Modul

```python
from PIL import Image, ImageEnhance
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import matplotlib.image as mpimg
from skimage import data
```

#1. Image brightening

```python
#read the image
im = Image.open("habibi.jpg")

#image brightness enhancer
enhancer = ImageEnhance.Brightness(im)
```

```python
factor1 = 1 #gives original image
img1 = enhancer.enhance(factor1)
img1
```

```python
factor2 = 0.5 #darkens the image
img2 = enhancer.enhance(factor2)
img2
```

```python
factor3 = 1.5 #brightens the image
img3 = enhancer.enhance(factor3)
img3
```

#2. Image Negatives

```python
img=mpimg.imread('habibi.jpg')
x=img[:,:,0]
y=np.shape(x)
z=np.zeros(y)
z=255-x
plt.xlabel("Value")
plt.ylabel("Pixel Frequency")
plt.title("Negative Image")
plt.imshow(z,cmap='gray')
plt.show()
```

#3. Contras enhancement

```python
image = data.camera()
```

```python
image_cs = np.zeros((image.shape[0],image.shape[1]),dtype = 'uint8')

min = np.min(image)
max = np.max(image)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image_cs[i,j] = 255*(image[i,j]-min)/(max-min)
```

```python
fig, axes = plt.subplots(5, 2, figsize=(20,20))
ax = axes.ravel()

ax[0].imshow(image,cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(image.ravel(),bins=256)
ax[1].set_title("Histogram Input")

ax[2].imshow(image_cs,cmap=plt.cm.gray)
ax[2].set_title("Citra Output CS")
ax[3].hist(image_cs.ravel(),bins=256)
ax[3].set_title("Histogram Output CS")

fig.tight_layout()
```

#4. Tresholding

```python
#membaca gambar bjhabibie 
img = cv2.imread('habibi.jpg',0)

# Hitungan threshold. 
# Perhatikan nilai ambang batas bawah dan atas dari tiap fungsi 
# yang diberikan
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

# menampilkan hasil
titles = ['Gambar asli','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# menampilkan beberapa gambar sekaligus
for i in range(6):
    # 3 baris, 2 kolom
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
